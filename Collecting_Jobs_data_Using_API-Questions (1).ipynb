You can also view  the json file contents  from the following <a href = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json">json</a> URL.


```python
# Step 1: Load the JSON data into a DataFrame
data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"
df = pd.read_json(data_url)

# Display the first few rows to ensure data is loaded correctly
print(df.head())
```

Calling the function for Python and checking if it works.


```python
def get_number_of_jobs_T(technology):
    # Filter the DataFrame to find jobs that mention the specified technology in 'Key Skills'
    filtered_df = df[df['Key Skills'].str.contains(technology, case=False, na=False)]
    
    # Count the number of job postings that match the filter
    number_of_jobs = filtered_df.shape[0]
    
    return technology, number_of_jobs

# Test the function for 'Python'
tech, count = get_number_of_jobs_T("Python")
print(f"Number of job postings for {tech}: {count}")
```

#### Write a function to find number of jobs in US for a location of your choice


```python
def get_number_of_jobs_L(location):
    # Filter the DataFrame to find jobs that match the specified location
    filtered_df = df[df['Location'].str.contains(location, case=False, na=False)]
    
    # Count the number of job postings that match the filter
    number_of_jobs = filtered_df.shape[0]
    
    return location, number_of_jobs
```

Call the function for Los Angeles and check if it is working.




```python
get_number_of_jobs_L("Los Angeles")
```

#### In the similar way, you can try for below given technologies and results  can be stored in an excel sheet.


Collect the number of job postings for the following languages using the API:

*   C
*   C#
*   C++
*   Java
*   JavaScript
*   Python
*   Scala
*   Oracle
*   SQL Server
*   MySQL Server
*   PostgreSQL
*   MongoDB


```python
import pandas as pd
import re

# List of technologies
technologies = [
    "C", "C#", "C++", "Java", "JavaScript", 
    "Python", "Scala", "Oracle", "SQL Server", 
    "MySQL Server", "PostgreSQL", "MongoDB"
]

# Initialize an empty list to store the results
tech_results = []

# Loop through each technology and get the number of job postings
for tech in technologies:
    # Escape special characters in the technology name
    escaped_tech = re.escape(tech)
    
    tech_name, count = get_number_of_jobs_T(escaped_tech)
    tech_results.append((tech_name, count))

# Convert the results list into a DataFrame
df_tech_results = pd.DataFrame(tech_results, columns=["Technology", "Number of Jobs"])

# Save the DataFrame to an Excel file named 'job-postings-by-technology.xlsx'
df_tech_results.to_excel('job-postings-by-technology.xlsx', index=False)

print("Data has been saved to 'job-postings-by-technology.xlsx'")
```
